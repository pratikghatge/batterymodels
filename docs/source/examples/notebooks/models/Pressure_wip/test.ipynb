```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 配置 Matplotlib 使用 Latin Modern 字体

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif"
})
# rcParams['font.family'] = 'serif'
# rcParams['font.serif'] = ['Latin Modern Roman']

# 示例数据
T_cell = [25, 50, 100, 150, 200]
pressure1 = [1, 2, 4, 7, 11]
pressure2 = [1, 1.5, 3, 6, 10]

# 创建图形
plt.plot(T_cell, pressure1, label='0.5 mol dm$^{-3}$', color='green')
plt.plot(T_cell, pressure2, label='1 mol dm$^{-3}$', color='orange')

# 添加标签和标题
plt.xlabel('T Cell / °C')
plt.ylabel('p Cell / bar')
plt.title('Pressure evolution with Latin Modern font')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
